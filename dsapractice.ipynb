```python
arr=[5,4,3,2,1]

def bubble_sort(arr):
    n=len(arr)
    for i in range(n):
        for j in range(0,n-i-1):
            if arr[j]>arr[j+1]:
                arr[j],arr[j+1]= arr[j+1],arr[j]
                print(i," :",arr)
        print(i,"finished loop")
m=bubble_sort(arr)
print(m)
```

```python
#selection sort

arr=[5,4,3,2,1]

def selection_sort(arr):
    n=len(arr)
    for i in range(n):
        lowest = i
        for j in range(i+1,n):
             if(arr[j]<arr[lowest]):
                lowest = j
        arr[i],arr[lowest] = arr[lowest],arr[i]
    return arr
            
m=selection_sort(arr) 
print(m)
```

```python
# insertion work own
arr = [5, 3, 4, 1, 2]

for i in range(1, len(arr)):
    key = arr[i]
    j = i - 1

    while j >= 0 and arr[j] > key:
        arr[j + 1] = arr[j]
        j -= 1

    arr[j + 1] = key

print(arr)
```

```python
arr = [2, 3, 4, 6, 7, 9]   # sorted array
search = 7

low = 0
high = len(arr) - 1

while low <= high:
    
    mid = (low + high) // 2

    if arr[mid] == search:
        print("Found at index", mid)
        break

    elif arr[mid] < search:
        low = mid + 1

    else:
        high = mid - 1

else:
    print("Element not found")
```
